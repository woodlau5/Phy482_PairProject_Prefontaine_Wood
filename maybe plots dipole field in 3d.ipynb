```python
from pylab import *
import numpy as np
from math import pi
mag_BB=7.965*10**15 #units of kg*m^3/(A*s^2)
dim = 3
def modsqr(A):
    size=len(A)
    xx=0
    for ix in range(0,size):
        xx=xx+A[ix]*A[ix]
    return xx
def dipole(x):
    BB=np.zeros(dim)
    r=np.sqrt(modsqr(x))
    BB[0]=(-mag_BB*3*x[0]*x[2]/(r**5))
    BB[1]=(-mag_BB*3*x[1]*x[2]/(r**5))
    BB[2]=(-mag_BB*(2*(x[2]**2)-(x[0]**2)-(x[1]**2))/(r**5))
    return BB
```

```python
fieldstrength = dipole([1,2,3])
print(fieldstrength)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

r=np.sqrt(modsqr([x,y,z]))

u=(-mag_BB*3*x*z/(r**5))
v=(-mag_BB*3*y*z/(r**5))
w=(-mag_BB*(2*(z**2)-(x**2)-(y**2))/(r**5))

ax.quiver(x, y, z, u, v, w, length=0.0001)

plt.show()
```
